Declaring vales

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('numeric.csv')

X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']
```

```python
def plot_vs(X, y, X_after, y_after, title, title_after):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot BEFORE
    df_before = pd.DataFrame(X.iloc[:, [1, 2]])  # radius2 and radius3
    df_before.columns = ['radius2', 'radius3']
    df_before['diagnosis'] = y
    sns.scatterplot(data=df_before, x='radius2', y='radius3', hue='diagnosis', palette='Set1', alpha=0.7, ax=axes[0])
    axes[0].set_title(title)
    axes[0].legend(title='Diagnosis')

    # Plot AFTER
    df_after = pd.DataFrame(X_after.iloc[:, [1, 2]])
    df_after.columns = ['radius2', 'radius3']
    df_after['diagnosis'] = y_after
    sns.scatterplot(data=df_after, x='radius2', y='radius3', hue='diagnosis', palette='Set1', alpha=0.7, ax=axes[1])
    axes[1].set_title(title_after)
    axes[1].legend(title='Diagnosis')

    plt.tight_layout()
    plt.show()
```

Applying Oversampling Techniques


Random Over Sampling

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler

#Random-Over-Sampling
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(X, y)
ros_df = pd.DataFrame(X_ros, columns = X.columns)
ros_df['diagnosis'] = y_ros

#Plot
plot_vs(X, y, ros_df, y_ros, 'Original', 'After ROS')
```

SMOTE

```python
#Reload dataset
df = pd.read_csv('numeric.csv')

X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']

#SMOTE
smote = SMOTE(random_state=42)
X_smoted, y_smoted = smote.fit_resample(X, y)
smoted_df = pd.DataFrame(X_smoted, columns = X.columns)
smoted_df['diagnosis'] = y_smoted

#Plot
plot_vs(X, y, smoted_df, y_smoted, 'Original', 'After SMOTE')
```

Applying Undersampling Techniques

```python
from imblearn.under_sampling import RandomUnderSampler, TomekLinks, NearMiss, CondensedNearestNeighbour, OneSidedSelection, EditedNearestNeighbours

#Reload dataset
df = pd.read_csv('numeric.csv')

X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']

#Random-Under-Sampling
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X, y)
rus_df = pd.DataFrame(X_rus, columns = X.columns)
rus_df['diagnosis'] = y_rus

#Plot
plot_vs(X, y, rus_df, y_rus, 'Original', 'After RUS')
```
