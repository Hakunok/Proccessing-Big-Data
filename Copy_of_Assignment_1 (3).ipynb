## Problem 1 - IsPalindrome - 15 points


Write a function to check if a given string is a palindrome and test it on the given input.

```python
def IsPalindrome(s: str) -> bool:
  # Update this code
  s = s.lower() # Lowercasing the string
  leg = len(s);
  for i in range(leg // 2):
    if s[i] != s[leg - i - 1]:
      return False
  return True
```

```python
# DO NOT MODIFY THIS CELL, BUT DO EXECUTE THIS CELL

a = ['', 'a', 'aa', 'abba', 'abab', 'abca']
expected = [True, True, True, True, False, False]

for ix, s in enumerate(a):
  b = IsPalindrome(s)
  print(s, '->IsPalindrom->', b)
  assert b == expected[ix]
```

# Problem 2 - FlipSort - 20 points

Use the function `FlipPrefix()` given below that can flip a prefix an array to sort the array in increasing order. You can compare array elements but the only way to modify the array is to use `FlipPrefix()`. You may want to use NumPy's `argmax()` function.

```python
# DO NOT MODIFY THIS CELL, BUT DO EXECUTE THIS CELL

def FlipPrefix(a: list[int], before_n: int) -> list[int]:
  assert 0 < before_n <= len(a)
  b = a[:before_n]
  return b[::-1] + a[before_n:]
```

```python
def FlipSort(a: list[int]) -> list[int]:
  # Update this code using FlipPrefix and argmax
  for i in range(len(a)):
    max = np.argmax(a[:len(a) - i])
    if max != 0:
      a = FlipPrefix(a, max + 1)

    a = FlipPrefix(a, len(a) - i)

  return a
```

```python
# DO NOT MODIFY THIS CELL, BUT DO EXECUTE THIS CELL

a = [[3, 1, 2, 5, 4],
     [1],
     [100, 101, 51, 41, 31, 21, 11, 1]]

for row in a:
  b = FlipSort(row)
  print(row, '->FlipSort->', b)
  assert b == sorted(row)
```

```python
import numpy as np
import matplotlib
import pandas as pd
from matplotlib import pyplot as plt

def initialize_W_and_B(features):
  w = np.zeros(features)
  # b = 0.0
  return w #, b

# prediction
def predict(X, w):
  return np.dot(X, w) # + b

def SGD(X_train, y_train, X_val, y_val, learning_rate, epochs):
  w = initialize_W_and_B(X_train.shape[1]) #, b
  samples = X_train.shape[0]
  train_losses = []
  val_losses = []

  for epoch in range(epochs):
    # Looping through the samples
    for i in range(samples):

      # predict outputs
      y_hat = predict(X_train[i], w) # + b

      # Update w
      w = w - learning_rate * ((y_hat - y_train[i])) *X_train[i]
      # b = b - learning_rate * (y_hat - y[i])

    # loss function for training set
    train_loss = np.sum((y_train - predict(X_train, w)) ** 2) / samples # - b in the sum
    train_losses.append(train_loss)

    # loss func for validation set
    y_Val_hat = predict(X_val, w)
    val_loss = np.sum((y_val - y_Val_hat) ** 2) / X_val.shape[0]
    val_losses.append(val_loss)

    # Printing training loss and validation loss side by side
    print(f"Epoch: {epoch:3d}, Training loss: {train_loss:.6f}    |    Validation loss: {val_loss:.6f}")

  return train_losses, val_losses, w

# Load training and testing data
train_data = pd.read_csv('train-1.csv')
test_data = pd.read_csv('test.csv')

# Splitting training and validating sets for 90:10 ratio
val_size = int(0.1 * len(train_data))
train_size = len(train_data) - val_size
train_data = train_data.sample(len(train_data))
train_data, val_data = train_data[:train_size], train_data[train_size:]

# Learning rate and epochs
learning_rate = 1e-9
epochs = train_size + val_size

# Training data
X_train = train_data.drop('arousal', axis = 1).values
y_train = train_data['arousal'].values

# Validating data
X_val = val_data.drop('arousal', axis = 1).values
y_val = val_data['arousal'].values

# Testing data
test_data = test_data.sample(len(test_data))
X_test = test_data.drop('arousal', axis = 1).values
y_test = test_data['arousal'].values

# Train the model + Validating model
train_loss, val_loss, w = SGD(X_train, y_train, X_val, y_val, learning_rate, epochs) #, b

# Test model
test_loss = np.sum((y_test - predict(X_test, w)) ** 2) / X_test.shape[0] #, b
print(f"\nTest set: {test_loss}")

#Plot the loss functions for training data, validation data
print("\n")
plt.plot(range(1, epochs + 1), train_loss, label = 'Training loss')
plt.plot(range(1, epochs + 1), val_loss, label = 'Validation loss')
#plt.plot(range(1, epochs + 1), w_test, label = 'Test loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.show()

```
